```python
%%writefile app.py
#Creamos el archivo de la APP en el interprete principal (Phyton)

#############################IMPLEMENTACIÓN DE DASHBOARD################################

#Verificamos que todas las librerias se puedan importar
#Importamos librerias
import streamlit as st
import plotly.express as px
import pandas as pd
import numpy as np
from funpymodeling.exploratory import freq_tbl
import folium
from streamlit_folium import folium_static
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import seaborn as sns
from io import BytesIO
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score


#################################################################

#Definimos la instancia de streamlit
@st.cache_resource

################################ CARGA DE ARCHIVOS #################################

#Creamos la función de carga de datos
def load_data():

  ####################### Extracción de características ###############################
    categoricas= pd.read_csv('Categoricas_rio.csv')
    categoricas.drop('Unnamed: 0', axis=1, inplace=True)
 ####################### REGRESION LINEAL ###############################

    coeficientes_hotel= pd.read_csv('Coeficientes_lineal_HOTEL_room.csv')
    coeficientes_hotel.rename(columns={'Unnamed: 0': 'Variables'}, inplace=True)
    coeficientes_hotel = coeficientes_hotel.sort_values(by='Coef_Correl', ascending=False)

    coeficientes_home= pd.read_csv('Coeficientes_lineal_home.csv')
    coeficientes_home.rename(columns={'Unnamed: 0': 'Variables'}, inplace=True)
    coeficientes_home = coeficientes_home.sort_values(by='Coef_Correl', ascending=False)

    coeficientes_private_room= pd.read_csv('Coeficientes_lineal_private_room.csv')
    coeficientes_private_room.rename(columns={'Unnamed: 0': 'Variables'}, inplace=True)
    coeficientes_private_room = coeficientes_home.sort_values(by='Coef_Correl', ascending=False)

    coeficientes_shared= pd.read_csv('Coeficientes_lineal_shared_room.csv')
    coeficientes_shared.rename(columns={'Unnamed: 0': 'Variables'}, inplace=True)
    coeficientes_shared = coeficientes_shared.sort_values(by='Coef_Correl', ascending=False)

  ####################### REGRESION NO LINEAL ###############################
    #Lectura del archivo csv sin indice
    data=pd.read_csv("rio_de_janeiro_USD.csv")
    data.drop(columns='Unnamed: 0', inplace=True)
    #Se crea una copia del frame original
    data_copia = data.copy()
    #Se detectan los valores unicos de la columna
    np.unique(data_copia['host_is_superhost'])

    #Se reemplazan los datos categoricos a numerico
    data_copia['host_is_superhost'] = data_copia['host_is_superhost'].replace('f',0)
    data_copia['host_is_superhost'] = data_copia['host_is_superhost'].replace('t',1)
    data_copia['host_is_superhost'] = data_copia['host_is_superhost'].astype(int)

    #Se detectan los valores unicos de la columna
    np.unique(data_copia['room_type'])


    #Se reemplazan los datos categoricos a numerico
    data_copia['room_type'] = data_copia['room_type'].replace('Entire home/apt',1)
    data_copia['room_type'] = data_copia['room_type'].replace('Hotel room',2)
    data_copia['room_type'] = data_copia['room_type'].replace('Private room',3)
    data_copia['room_type'] = data_copia['room_type'].replace('Shared room',4)
    data_copia['room_type'] = data_copia['room_type'].astype(int)

    #Se eliminan columnas innecesarias
    data_evaluar = data_copia.select_dtypes(include=[int, float])
    data_evaluar.head(3)
    #Encontramos todos los coeficientes de correlacion entre las variables
    corr_factors = data_evaluar.corr()
    #Encontramos el valor absoluto de todas las correlaciones entre las variables
    corr_factors1 = abs(corr_factors)

    data_evaluar2= data_evaluar.copy()

    #Selecciono las columnas tipo numericas del dataframe Filtro_index1
    numeric_df1 = data_evaluar.select_dtypes(['float','int'])  #Devuelve Columnas
    numeric_cols1= numeric_df1.columns                         #Devuelve lista de Columnas

    numeric2_df1=data_evaluar[['host_response_rate','host_acceptance_rate','host_total_listings_count','accommodates','reviews_per_month','price']]

    top_modelos_correlacion= pd. read_csv('k.csv', index_col='Variable')

  ########################## RETURN DE LA FUNCIÓN ##################################
    return data_evaluar, numeric_df1, numeric_cols1, corr_factors, corr_factors1, numeric2_df1, top_modelos_correlacion, categoricas, coeficientes_home,coeficientes_hotel, coeficientes_private_room, coeficientes_shared, data_evaluar2

#################################################################

#Cargo los datos obtenidos de la función "load_data"
data_evaluar, numeric_df1, numeric_cols1, corr_factors, corr_factors1, numeric2_df1, top_modelos_correlacion, categoricas, coeficientes_home,coeficientes_hotel, coeficientes_private_room, coeficientes_shared, data_evaluar2= load_data()

######################CREACIÓN DEL DASHBOARD################################

#1. CREACIÓN DE LA SIDEBAR
#Generamos los encabezados para la barra lateral (sidebar)
st.sidebar.title("Airbnb RIO DE JANEIRO César Alejandro Rivera Guzmán")
# URL of the image
image = "https://i.pinimg.com/originals/3c/bf/be/3cbfbe148597341fa56f2f87ade90956.png"
st.sidebar.image(image, use_column_width=True)
st.sidebar.header("Act 3.7 Despliegue de Datos")
st.sidebar.header('SIDEBAR')
st.sidebar.header("Panel de selección")



#Test 2 (streamlit run app.py)

#################################################################

#2. CREACIÓN DE LOS FRAMES
#Generamos los Frames que utilizaremos en el diseño
#Widget 1: Selectbox
#Menu desplegable de opciones de los Frames seleccionados
Frames= st.selectbox(label= "Frames", options= ["Descripción de la actividad",'Rio de Janeiro','Extracción de Características','Regresión Lineal','Regresión Multiple',"Regresión NO Lineal",'Regresión Logística'])

#Test 3 (streamlit run app.py)

############################# FRAME 1 ####################################

if Frames == 'Descripción de la actividad':
    st.title('Encontrar el mejor modelo de predicción para la variables estudio que son las siguientes:')
    text = """
    ### Etapa I. Modelado explicativo

    **a) Extracción de características.** Análisis univariado de las variables categóricas más significativas.

    ### Etapa II. Modelado predictivo

    **b) Análisis de correlaciones.** Análisis aplicando los distintos tipos de regresión vistos en el curso:
    - Regresión lineal simple
    - Regresión lineal múltiple
    - Regresión no lineal
    - Regresión logística

    ### Desarrollo

    - Desarrollar como mínimo 2 **FRAMES** por cada ciudad (Evaluación Individual por ciudad).

    ### Dashboard

    - Agregar al menos 1 vez los siguientes **widgets** al **DASHBOARD**:
        - Select box
        - Check box
        - Multiselect box
        - Button
        - Lineplot
        - Barplot
        - Scatterplot
        - Pieplot

    ### Bonus Extra
    En la evaluación se considerará como un **bonus extra** la modificación del diseño y la implementación de widgets adicionales a los que se presenten en la Plantilla descrita en la clase, por ejemplo:
        - Heatmap
        - Boxplot
        - Tablas
        - Selectores
        - Iconos
        - Color, etc.
    """
    st.write(text)

######################### FRAME 2 ########################################
if Frames == "Rio de Janeiro":
    # Generamos los encabezados para el dashboard
    st.title("**Rio de Janeiro**")
    st.header("Información de Rio de Janeiro")
    info = """
    **RIO DE JANEIRO - INFORMACIÓN GENERAL**

    **1. GEOGRAFÍA**
    - **Ubicación**: Sureste de Brasil, en la costa atlántica.
    - **Población**: Aproximadamente 6.7 millones de habitantes (2023).
    - **Superficie**: Alrededor de 1,200 km².
    - **Clima**: Tropical atlántico, con veranos cálidos y húmedos e inviernos suaves.

    **2. HISTORIA**
    - **Fundación**: 1 de marzo de 1565, por los portugueses.
    - **Capital**: Fue la capital de Brasil hasta 1960.
    - **Desarrollo histórico**: Ha sido un importante puerto comercial y un centro clave en la política y cultura brasileñas.

    **3. ECONOMÍA**
    - **Moneda de Brasil**: Real brasileño (BRL).
    - **Industrias clave**: Petróleo, turismo, comercio, finanzas, entretenimiento y medios.
    - **Turismo**: Destino turístico popular por sus playas (Copacabana, Ipanema), el Cristo Redentor, el Pan de Azúcar, y el Carnaval de Río.
    - **Infraestructura**: Importante puerto y aeropuerto internacional (Galeão).

    **4. CULTURA**
    - **Carnaval**: Uno de los más grandes y famosos del mundo, con desfiles de samba y fiestas que atraen a millones de turistas.
    - **Música**: Cuna de la samba y la bossa nova.
    - **Patrimonio mundial**: Declarado Patrimonio de la Humanidad por la UNESCO en 2012 por su combinación única de paisajes naturales y urbanismo.
    - **Fútbol**: Sede del Estadio Maracaná, uno de los más emblemáticos del mundo.
    """
    st.write(info)

    # Agregar la imagen RIO DE JANEIRO
    image_RIO_url = "https://media.architecturaldigest.com/photos/56f4781d904aa1ac0ac47256/16:9/w_2580,c_limit/Cristo%20Redentor.jpg"
    st.image(image_RIO_url, caption="Imagen de Rio de Janeiro", use_column_width=True)

    image_AIRBNB_url = 'https://brandemia.org/sites/default/files/sites/default/files/bnb_billboard_01-2000x1125.jpg'
    st.image(image_AIRBNB_url, caption="Airbnb", use_column_width=True)

    # Crear un mapa centrado en un lugar específico
    m = folium.Map(location=[-22.908333, -43.196388], zoom_start=5)

    # Agregar un marcador
    folium.Marker([-22.908333, -43.196388], popup="Rio de Janeiro, BRA").add_to(m)

    # Mostrar el mapa
    folium_static(m)

######################### FRAME 3 ########################################
if Frames== ('Extracción de Características'):
    st.title('**Extracción de Características**')
    st.subheader('*Análisis Univariado*')
    texto_instrucciones = """

    1. **Extraer características** a partir de aplicar un **análisis univariado** a diferentes variables categóricas (20). Mostrar los resultados obtenidos mediante **gráficas y tablas**.

    2. **Categorizar las siguientes variables**:
        - “host_response_rate”
        - “host_acceptance_rate”
        - “host_total_listings_count”
        - “accommodates”
        - "bathrooms"
        - “beds”
        - “price”
        - “maximum_nights_avg_ntm”
        - “availability_365”
        - “number_of_reviews”
        - “review_scores_value”
        - “reviews_per_month”

    3. Aplicar la **regla de Sturges** y desplegar en un gráfico las frecuencias de las categorías obtenidas.
    """
    st.write(texto_instrucciones)

    Vars_num_selected3 = st.sidebar.multiselect(
        label="Variable a categorizar",
        options=[
            'host_response_rate',
            'host_acceptance_rate',
            'host_total_listings_count',
            'accommodates',
            'bathrooms',
            'beds',
            'price',
            'maximum_nights_avg_ntm',
            'availability_365',
            'number_of_reviews',
            'review_scores_value',
            'reviews_per_month'
        ]
    )
    if Vars_num_selected3:
        for variable in Vars_num_selected3:
          # Obtengo un análisis univariado de una variable categórica en específico
          table1 = freq_tbl(categoricas[variable])
          st.title(f'**Análisis Univariado** variable: **{variable}**')
          st.write(table1)

          # Elimino las columnas "percentage" y "cumulative_perc" para trabajar con las frecuencias
          table2 = table1.drop(['percentage', 'cumulative_perc'], axis=1)

          # Obtengo un filtro de los valores más relevantes de la variable categórica seleccionada
          Filtro = table2[table2['frequency'] > 1]

          # Ajusto el índice de mi DataFrame
          Filtro_index = Filtro.set_index(variable)

          # Crear el gráfico de barras interactivo
          bar_fig = px.bar(Filtro_index, x=Filtro_index.index, y='frequency',
                            title=f'{variable} vs frecuencias',
                            labels={'frequency': 'Frecuencia', variable: f'Rango de {variable}'})
          st.plotly_chart(bar_fig)

          # Crear el gráfico de dispersión interactivo
          scatter_fig = px.scatter(table1, x='frequency', y='cumulative_perc',
                                    title=f'Gráfico de dispersión de {variable} frecuencia vs porcentaje acumulado',
                                    labels={'frequency': 'Frecuencia', 'cumulative_perc': 'Porcentaje acumulado'})
          st.plotly_chart(scatter_fig)

          # Crear el gráfico de pastel interactivo
          pie_fig = px.pie(Filtro_index, values='frequency', names=Filtro_index.index,
                          title='Frecuencia de categorías',
                          hover_data=['frequency'],
                          labels={'frequency': 'Frecuencia'})
          st.plotly_chart(pie_fig)

######################### FRAME 4 ########################################
if Frames == "Regresión Lineal":
    st.title("*Regresiones lineal*")
    st.subheader('### Análisis de Correlación y Modelado Matemático')
    text = """

    1. **Analizar la correlación** que existe en cada tipo de habitación (**room_type**), eligiendo 4 tipos, respecto a las siguientes variables (en el orden dependiente, independiente)
        - **“host_acceptance_rate vs host_response_rate”**
        - **“review_scores_location vs review_scores_cleanliness”**
        - **“host_acceptance_rate vs price”**
        - **“availability_365 vs number_of_reviews”**
        - **“host_acceptance_rate vs number_of_reviews”**
        - **“reviews_per_month vs review_scores_communication”**
    """
    st.write(text)

    st.write('*NOTA, el tipo de cuarto se denominó de la siguiente forma:\
    \n1 Entire home/apt\
    \n2 Hotel room\
    \n3 Private room\
    \n4 Shared room')

    room_type=st.sidebar.multiselect(
        label="Tipo de cuarto",
      options=[
            1,
            2,
            3,
            4])

    if room_type:
      for tipo_cuarto in room_type:
        cuarto = data_evaluar[data_evaluar['room_type'] == tipo_cuarto]
        st.title(f'Tipo de cuarto: {tipo_cuarto}')
        st.write(f'Se cuenta con un total de **{len(cuarto)}**')

        Vars_num_selected7 = st.sidebar.multiselect(
            label="Variables",
          options=[
                'host_acceptance_rate vs host_response_rate',
                'review_scores_location vs review_scores_cleanliness”',
                'host_acceptance_rate vs price',
                'availability_365 vs number_of_reviews',
                'host_acceptance_rate vs number_of_reviews'])

      if 'host_acceptance_rate vs host_response_rate' in Vars_num_selected7:

        st.title('host_acceptance_rate vs host_response_rate')
        #Declaramos las variables dependientes e independientes para la regresion lineal
        vars_indep_A1 = cuarto[['host_response_rate']]
        var_dep_A1 = cuarto['host_acceptance_rate']
        model_A1 = LinearRegression()
        #Ajustamos el modelo con las variables antes declaradas
        model_A1.fit(X=vars_indep_A1, y=var_dep_A1)
        #Predecimos los valores a partir de la variable
        y_pred_A1 = model_A1.predict(X=cuarto[['host_response_rate']])
        #Insertamos la columna de predicciones en el DataFrame
        cuarto.insert(0,'predicciones_1',y_pred_A1)
        # Crea la figura y el eje
        fig, ax = plt.subplots(figsize=(10, 6))

        # Crea la gráfica de dispersión para host_response_rate y host_acceptance_rate
        sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color='blue', data=cuarto, ax=ax, label='host_acceptance_rate')

        # Crea la gráfica de dispersión para host_response_rate y predicciones_1
        sns.scatterplot(x='host_response_rate', y='predicciones_1', color='red', data=cuarto, ax=ax, label='predicciones_1')

        # Agrega título y etiquetas a los ejes
        ax.set_title('Gráfica comparativa')
        ax.set_xlabel('host_response_rate')
        ax.set_ylabel('Valor')

        # Agrega leyenda
        ax.legend()

        # Muestra la gráfica en Streamlit
        st.pyplot(fig)

        #Corrobaremos cual es el coeficiente de Determinacion de nuestro modelo (R_Cuadrada)
        coef_deter_A1 = model_A1.score(X=vars_indep_A1, y=var_dep_A1)

        st.write(f'Modelo matemático:  {model_A1.__dict__}')

        st.write(f'Coeficiente de Determinacion de nuestro modelo (R_Cuadrada): {coef_deter_A1}')

        #Corroboramos cual es el coeficiente de correlacion de nuestro modelo (R)
        coef_correl_A1 = np.sqrt(coef_deter_A1)
        st.write(f'Coeficiente de Correlacion de nuestro modelo (R): {coef_correl_A1}')

      elif 'review_scores_location vs review_scores_cleanliness' in Vars_num_selected7:
        st.title('review_scores_location vs review_scores_cleanliness')
        #Declaramos las variables dependientes e independientes para la regresion lineal
        vars_indep_A1 = cuarto[['review_scores_cleanliness']]
        var_dep_A1 = cuarto['review_scores_location']
        model_A1 = LinearRegression()
        #Ajustamos el modelo con las variables antes declaradas
        model_A1.fit(X=vars_indep_A1, y=var_dep_A1)
        #Predecimos los valores a partir de la variable
        y_pred_A1 = model_A1.predict(X=cuarto[['review_scores_cleanliness']])
        #Insertamos la columna de predicciones en el DataFrame
        cuarto.insert(1,'predicciones_2',y_pred_A1)
        # Crea la figura y el eje
        fig, ax = plt.subplots(figsize=(10, 6))

        # Crea la gráfica de dispersión para host_response_rate y host_acceptance_rate
        sns.scatterplot(x='review_scores_cleanliness', y='review_scores_location', color='blue', data=cuarto, ax=ax, label='review_scores_cleanliness')

        # Crea la gráfica de dispersión para host_response_rate y predicciones_1
        sns.scatterplot(x='review_scores_cleanliness', y='predicciones_2', color='red', data=cuarto, ax=ax, label='predicciones_2')

        # Agrega título y etiquetas a los ejes
        ax.set_title('Gráfica comparativa')
        ax.set_xlabel('review_scores_cleanliness')
        ax.set_ylabel('Valor')

        # Agrega leyenda
        ax.legend()

        # Muestra la gráfica en Streamlit
        st.pyplot(fig)

        #Corrobaremos cual es el coeficiente de Determinacion de nuestro modelo (R_Cuadrada)
        coef_deter_A1 = model_A1.score(X=vars_indep_A1, y=var_dep_A1)

        st.write(f'Modelo matemático:  {model_A1.__dict__}')

        st.write(f'Coeficiente de Determinacion de nuestro modelo (R_Cuadrada): {coef_deter_A1}')

        #Corroboramos cual es el coeficiente de correlacion de nuestro modelo (R)
        coef_correl_A1 = np.sqrt(coef_deter_A1)
        st.write(f'Coeficiente de Correlacion de nuestro modelo (R): {coef_correl_A1}')

      elif 'host_acceptance_rate vs price' in Vars_num_selected7:
        st.title('host_acceptance_rate vs price')
        #Declaramos las variables dependientes e independientes para la regresion lineal
        vars_indep_A1 = cuarto[['price']]
        var_dep_A1 = cuarto['host_acceptance_rate']
        model_A1 = LinearRegression()
        #Ajustamos el modelo con las variables antes declaradas
        model_A1.fit(X=vars_indep_A1, y=var_dep_A1)
        #Predecimos los valores a partir de la variable
        y_pred_A1 = model_A1.predict(X=cuarto[['price']])
        #Insertamos la columna de predicciones en el DataFrame
        cuarto.insert(2,'predicciones_3',y_pred_A1)
        # Crea la figura y el eje
        fig, ax = plt.subplots(figsize=(10, 6))

        # Crea la gráfica de dispersión para host_response_rate y host_acceptance_rate
        sns.scatterplot(x='price', y='review_scores_location', color='blue', data=cuarto, ax=ax, label='price')

        # Crea la gráfica de dispersión para host_response_rate y predicciones_1
        sns.scatterplot(x='price', y='predicciones_3', color='red', data=cuarto, ax=ax, label='predicciones_3')

        # Agrega título y etiquetas a los ejes
        ax.set_title('Gráfica comparativa')
        ax.set_xlabel('price')
        ax.set_ylabel('Valor')

        # Agrega leyenda
        ax.legend()

        # Muestra la gráfica en Streamlit
        st.pyplot(fig)

        #Corrobaremos cual es el coeficiente de Determinacion de nuestro modelo (R_Cuadrada)
        coef_deter_A1 = model_A1.score(X=vars_indep_A1, y=var_dep_A1)

        st.write(f'Modelo matemático:  {model_A1.__dict__}')

        st.write(f'Coeficiente de Determinacion de nuestro modelo (R_Cuadrada): {coef_deter_A1}')

        #Corroboramos cual es el coeficiente de correlacion de nuestro modelo (R)
        coef_correl_A1 = np.sqrt(coef_deter_A1)
        st.write(f'Coeficiente de Correlacion de nuestro modelo (R): {coef_correl_A1}')

      elif 'availability_365 vs number_of_reviews' in Vars_num_selected7:
        st.title('availability_365 vs number_of_reviews')
        #Declaramos las variables dependientes e independientes para la regresion lineal
        vars_indep_A1 = cuarto[['number_of_reviews']]
        var_dep_A1 = cuarto['availability_365']
        model_A1 = LinearRegression()
        #Ajustamos el modelo con las variables antes declaradas
        model_A1.fit(X=vars_indep_A1, y=var_dep_A1)
        #Predecimos los valores a partir de la variable
        y_pred_A1 = model_A1.predict(X=cuarto[['number_of_reviews']])
        #Insertamos la columna de predicciones en el DataFrame
        cuarto.insert(3,'predicciones_4',y_pred_A1)
        # Crea la figura y el eje
        fig, ax = plt.subplots(figsize=(10, 6))

        # Crea la gráfica de dispersión para host_response_rate y host_acceptance_rate
        sns.scatterplot(x='number_of_reviews', y='availability_365', color='blue', data=cuarto, ax=ax, label='number_of_reviews')

        # Crea la gráfica de dispersión para host_response_rate y predicciones_1
        sns.scatterplot(x='number_of_reviews', y='predicciones_4', color='red', data=cuarto, ax=ax, label='predicciones_4')

        # Agrega título y etiquetas a los ejes
        ax.set_title('Gráfica comparativa')
        ax.set_xlabel('number_of_reviews')
        ax.set_ylabel('Valor')

        # Agrega leyenda
        ax.legend()

        # Muestra la gráfica en Streamlit
        st.pyplot(fig)

        #Corrobaremos cual es el coeficiente de Determinacion de nuestro modelo (R_Cuadrada)
        coef_deter_A1 = model_A1.score(X=vars_indep_A1, y=var_dep_A1)

        st.write(f'Modelo matemático:  {model_A1.__dict__}')

        st.write(f'Coeficiente de Determinacion de nuestro modelo (R_Cuadrada): {coef_deter_A1}')

        #Corroboramos cual es el coeficiente de correlacion de nuestro modelo (R)
        coef_correl_A1 = np.sqrt(coef_deter_A1)
        st.write(f'Coeficiente de Correlacion de nuestro modelo (R): {coef_correl_A1}')

      elif 'host_acceptance_rate vs number_of_reviews' in Vars_num_selected7:
        st.title('host_acceptance_rate vs number_of_reviews')
        #Declaramos las variables dependientes e independientes para la regresion lineal
        vars_indep_A1 = cuarto[['number_of_reviews']]
        var_dep_A1 = cuarto['host_acceptance_rate']
        model_A1 = LinearRegression()
        #Ajustamos el modelo con las variables antes declaradas
        model_A1.fit(X=vars_indep_A1, y=var_dep_A1)
        #Predecimos los valores a partir de la variable
        y_pred_A1 = model_A1.predict(X=cuarto[['number_of_reviews']])
        #Insertamos la columna de predicciones en el DataFrame
        cuarto.insert(4,'predicciones_5',y_pred_A1)
        # Crea la figura y el eje
        fig, ax = plt.subplots(figsize=(10, 6))

        # Crea la gráfica de dispersión para host_response_rate y host_acceptance_rate
        sns.scatterplot(x='number_of_reviews', y='availability_365', color='blue', data=cuarto, ax=ax, label='number_of_reviews')

        # Crea la gráfica de dispersión para host_response_rate y predicciones_1
        sns.scatterplot(x='number_of_reviews', y='predicciones_5', color='red', data=cuarto, ax=ax, label='predicciones_5')

        # Agrega título y etiquetas a los ejes
        ax.set_title('Gráfica comparativa')
        ax.set_xlabel('number_of_reviews')
        ax.set_ylabel('Valor')

        # Agrega leyenda
        ax.legend()

        # Muestra la gráfica en Streamlit
        st.pyplot(fig)

        #Corrobaremos cual es el coeficiente de Determinacion de nuestro modelo (R_Cuadrada)
        coef_deter_A1 = model_A1.score(X=vars_indep_A1, y=var_dep_A1)

        st.write(f'Modelo matemático:  {model_A1.__dict__}')

        st.write(f'Coeficiente de Determinacion de nuestro modelo (R_Cuadrada): {coef_deter_A1}')

        #Corroboramos cual es el coeficiente de correlacion de nuestro modelo (R)
        coef_correl_A1 = np.sqrt(coef_deter_A1)
        st.write(f'Coeficiente de Correlacion de nuestro modelo (R): {coef_correl_A1}')

######################### FRAME 5 ########################################
if Frames == "Regresión Multiple":
    st.title("*Regresiones NO lineales*")
    g= '''
    1. **Crear el mejor modelo de regresión lineal múltiple** para cada variable cuantitativa
        - **host_acceptance_rate**
        - **host_is_superhost**
        - **host_total_listings_count**
        - **accommodates**
        - **bedrooms**
        - **price**
        - **review_scores_value**
        - **reviews_per_month**
    '''
    st.write(g)

    st.header('Mejores modelos tipo de cuarto **HOTEL**')
    st.write(coeficientes_hotel)

    st.header('Mejores modelos tipo de cuarto **HOME**')
    st.write(coeficientes_home)

    st.header('Mejores modelos tipo de cuarto **Private Room**')
    st.write(coeficientes_private_room)

    st.header('Mejores modelos tipo de cuarto **Shared Room**')
    st.write(coeficientes_shared)

    st.write('*NOTA, el tipo de cuarto se denominó de la siguiente forma:\
    \n1 Entire home/apt\
    \n2 Hotel room\
    \n3 Private room\
    \n4 Shared room')

    room_types=st.sidebar.multiselect(
        label="Tipo de cuarto",
      options=[
            1,
            2,
            3,
            4])
        #Widget 2: Checkbox
    check_box = st.sidebar.checkbox(label="Mostrar Mapa de Calor")
    if check_box:
      def HeatMap():
        plt.figure(figsize=(10, 8))  # Set the figure size
        Heat_Map = sns.heatmap(corr_factors1, cmap='Blues')
        plt.title('Mapa de calor Correlaciones Regresión NO lineal')
        st.pyplot(plt)  # Use st.pyplot to display the plot in Streamlit

      st.title('**Heat Map**')
      HeatMap()  # Call the HeatMap function
        #Mostramos el HEATMAP

    if room_types:
      for tipo_cuartos in room_types:
          cuarto = data_evaluar[data_evaluar['room_type'] == tipo_cuartos]

          variables = st.sidebar.multiselect(
              label="Variables a evaluar",
              options=[
                  'host_response_rate',
                  'host_acceptance_rate',
                  'host_total_listings_count',
                  'accommodates',
                  'reviews_per_month',
                  'price',
                  'host_is_superhost',
                  'bedrooms',
                  'review_scores_value'
              ]
          )
          if variables:
              for vari in variables:
                  a = corr_factors1.sort_values(by=vari, ascending=False).head(7)
                  a = a[1:].head()
                  st.write(f'Tabla Correlación para {vari}')
                  st.write(a)

                  # Variable Independiente
                  a1 = a.index.tolist()  # Asegúrate de que a1 sea una lista
                  st.write('Variables dependientes:', a1)

                  # Declaramos las variables dependientes e independientes para la regresión lineal
                  vars_indep_1 = data_evaluar2[a1]  # Usa a1 directamente como lista
                  var_dep_1 = data_evaluar2[vari]

                  model_1 = LinearRegression()

                  # Ajustamos el modelo con las variables antes declaradas
                  model_1.fit(X=vars_indep_1, y=var_dep_1)

                  # Predecimos los valores
                  y_pred_1 = model_1.predict(X=vars_indep_1)  # Usa vars_indep_1 para predecir

                  # Insertamos la columna de predicciones en el DataFrame
                  data_evaluar2.insert(0, 'predicciones', y_pred_1)

                  # Crea la figura y el eje
                  fig, ax = plt.subplots(figsize=(10, 6))

                  # Visualiza la gráfica comparativa
                  sns.scatterplot(x=a1[0], y=vari, color='blue', data=data_evaluar, ax=ax, label='Total Real')
                  sns.scatterplot(x=a1[0], y='predicciones', color='red', data=data_evaluar2, ax=ax, label='Predicciones')  # Cambia 'predicciones_1' por 'predicciones'

                  # Agrega título y etiquetas a los ejes
                  ax.set_title(f'{a1[0]} vs {vari} / Regresión Múltiple')
                  ax.set_xlabel(a1[0])
                  ax.set_ylabel('Valor')

                  # Agrega leyenda
                  ax.legend()

                  # Muestra la gráfica en Streamlit
                  st.pyplot(fig)

                  # Coeficiente de Determinación
                  coef_deter_1 = model_1.score(X=vars_indep_1, y=var_dep_1)

                  # Calculamos el coeficiente de determinación del modelo
                  R2_Modelo14 = r2_score(var_dep_1, y_pred_1)
                  st.write(f'Coeficiente de determinación del modelo {vari}: {R2_Modelo14}')

                  R_14 = np.sqrt(R2_Modelo14)
                  st.write(f'Coeficiente de correlación del modelo {vari}: {R_14}')




######################### FRAME 6 ########################################
if Frames == "Regresión NO Lineal":
    st.title("*Regresiones NO lineales*")
    st.subheader('Encontrar el mejor modelo de predicción para la variables estudio que son las siguientes:')
    st.write('"Para esta sección serán las siguientes: : \
  \n-host_response_rate (variable objetivo)\
  \n-host_acceptance_rate (variable objetivo)\
  \n-host_total_listings_count (variable objetivo)\
  \n-accommodates (variable objetivo)\
  \n-reviews_per_month (variable objetivo)\
  \n-price(variable objetivo)')

    #Widget 2: Checkbox
    check_box = st.sidebar.checkbox(label="Mostrar Mapa de Calor")
    if check_box:
      def HeatMap():
        plt.figure(figsize=(10, 8))  # Set the figure size
        Heat_Map = sns.heatmap(corr_factors1, cmap='Blues')
        plt.title('Mapa de calor Correlaciones Regresión NO lineal')
        st.pyplot(plt)  # Use st.pyplot to display the plot in Streamlit

      st.title('**Heat Map**')
      HeatMap()  # Call the HeatMap function
        #Mostramos el HEATMAP

    st.title('**Modelos** utilizados:')
    st.write('\
    \n-Modelo Exponencial\
    \n-Modelo Función Inversa\
    \n-Modelo Senoidal\
    \n-Modelo Tangencial\
    \n-Modelo Valor Absoluto\
    \n-Modelo Función Cociente entre Polinomios\
    \n-Modelo Logarítmica\
    \n-Modelo Función Lineal con Producto de Coeficientes\
    \n-Modelo Cuadrático Inverso\
    \nModelo Polinomial Inverso')

    #Widget 2: Checkbox
    check_box2 = st.sidebar.checkbox(label="TOP Modelos de Correlación Variables Estudio")
    if check_box2:
      def Corr_Det():
        st.title('TOP Modelos de Correlación Variables Estudio')
        st.write(top_modelos_correlacion)

      Corr_Det()  # Call the HeatMap function
        #Mostramos el HEATMAP

    # Generamos un cuadro de multi-selección (Y) para seleccionar variables a graficar
    Vars_num_selected = st.sidebar.multiselect(
        label="Variable dependiente de mayor coeficiente de correlación",
        options=[
            'host_response_rate',
            'host_acceptance_rate',
            'host_total_listings_count',
            'accommodates',
            'reviews_per_month',
            'price'
        ]
    )

    if Vars_num_selected:
        for var in Vars_num_selected:
            a = corr_factors1.sort_values(by=var, ascending=False).head(7)
            a = a[1:].head()
            st.write(f'Tabla Correlación para {var}')
            st.write(a)
            # Variable Independiente
            a1 = a.index[0]
            st.write('Variable dependiente:', a1)

            #Declaramos las variables dependientes e independientes para la regresión lineal
            vars_indep = data_evaluar[a1]
            var_dep = data_evaluar[var]
            x=vars_indep
            y=var_dep
                # Generamos un cuadro de multi-selección (Y) para seleccionar variables a graficar
            Vars_num_selected2 = st.sidebar.multiselect(
            label="Modelo de Regresión: ",
            options=[
            'Modelo Cuadrático',
            'Modelo Exponencial',
            'Modelo Función Inversa',
            'Modelo Senoidal',
            'Modelo Tangencial',
            'Modelo Valor Absoluto',
            'Modelo Función Cociente entre Polinomios',
            'Modelo Logarítmica',
            'Modelo Función Lineal con Producto de Coeficientes',
            'Modelo Cuadrático Inverso'
            'Modelo Polinomial Inverso'
        ]
    )
            if 'Modelo Cuadrático' in Vars_num_selected2:
              try:
                def func_Cuadratica(x, a, b, c):
                  return a * np.square(x) + b * x + c

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_Cuadratica, vars_indep, var_dep)

                # Predicciones
                y_pred = func_Cuadratica(vars_indep, *parametros)

                st.title('Modelo Cuadrático')

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Exponencial' in Vars_num_selected2:
              try:
                def func_Expo(x, a, b, c):
                  return a * np.exp(-b * x) + c

                st.title('Modelo Exponencial')

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_Expo, vars_indep, var_dep)

                # Predicciones
                y_pred = func_Expo(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Función Inversa' in Vars_num_selected2:
              try:
                def func_inv(x, a, b, c):
                  return 1/a*x

                st.title('Modelo Función Inversa')
                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_inv, vars_indep, var_dep)

                # Predicciones
                y_pred = func_inv(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Senoidal' in Vars_num_selected2:
              try:
                def func_sen(x, a, b, c):
                  return a*np.sin(x) + b

                st.title('Modelo Senoidal')
                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_sen, vars_indep, var_dep)

                # Predicciones
                y_pred = func_sen(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Tangencial' in Vars_num_selected2:
              try:
                def func_tan(x, a, b, c):
                  return a*np.tan(x) + b

                st.title('Modelo Tangencial')
                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_tan, vars_indep, var_dep)

                # Predicciones
                y_pred = func_tan(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Valor Absoluto' in Vars_num_selected2:
              try:
                def func_abs(x, a, b, c):
                  return a*np.abs(x) + b*x + c
                st.title('Modelo Valor Absoluto')

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_abs, vars_indep, var_dep)

                # Predicciones
                y_pred = func_abs(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Función Cociente entre Polinomios' in Vars_num_selected2:
              try:
                def func_coci_poli(x, a, b, c):
                  return (a*x**2 + b)/ c*x
                st.title('Modelo Función Cociente entre Polinomios')

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_coci_poli, vars_indep, var_dep)

                # Predicciones
                y_pred = func_coci_poli(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Logarítmica' in Vars_num_selected2:
              try:
                def func_log(x, a, b, c):
                  return a*np.log(x) + b
                st.title("Modelo Logarítmica")

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_log, vars_indep, var_dep)

                # Predicciones
                y_pred = func_log(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Función Lineal con Producto de Coeficientes' in Vars_num_selected2:
              try:
                def func_lin_prod(x, a, b, c):
                  return a*x + b*x + c*x

                st.title("Modelo Función Lineal con Producto de Coeficientes")

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_lin_prod, vars_indep, var_dep)

                # Predicciones
                y_pred = func_lin_prod(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Cuadrático Inverso' in Vars_num_selected2:
              try:
                def func_cuad_inv(x, a, b, c):
                  return 1/a*x**2

                st.title("Modelo Cuadrático Inverso")

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_cuad_inv, vars_indep, var_dep)

                # Predicciones
                y_pred = func_cuad_inv(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")

            elif 'Modelo Polinomial Inverso' in Vars_num_selected2:
              try:
                def func_poli_inv(x, a, b, c):
                  return a/b*x**2 + c*x

                st.title("Modelo Polinomial Inverso")

                # Ajustamos los parámetros de la función curve_fit
                parametros, covs = curve_fit(func_poli_inv, vars_indep, var_dep)

                # Predicciones
                y_pred = func_poli_inv(vars_indep, *parametros)

                # Calculamos el coeficiente de determinación del modelo
                R2_Modelo14 = r2_score(var_dep, y_pred)
                st.write(f'Coeficiente de determinacion del modelo {var}:')
                st.write(R2_Modelo14)

                # Calculamos el coeficiente de correlación del modelo
                R_14 = np.sqrt(R2_Modelo14)
                st.write(f'Coeficiente de correlación del modelo {var}:')
                st.write(R_14)

                figure1 = px.scatter(data_frame=data_evaluar,
                        x=vars_indep,
                        y=[var_dep, y_pred],
                        title=str(var),
                        labels={'x': 'Valor de la variable independiente',
                                'y': 'Valor de la variable dependiente'},
                        width=1600, height=600,
                        color_discrete_sequence=['blue', 'red'],
                        color_discrete_map={'Valores reales': 'blue', 'Valores predecidos': 'red'})

                st.plotly_chart(figure1)
              except Exception as e:
                st.write(f"An unexpected error occurred: {e}")




###################### FRAME 7########################
if Frames == "Regresión Logística":
    st.title("*Regresiones Logística*")

    variables = st.sidebar.multiselect(
              label="Variables a evaluar",
              options= numeric_cols1)
    if variables:
      try:
        for vari in variables:
          a = corr_factors1.sort_values(by=vari, ascending=False).head(7)
          a = a[1:].head(4)
          st.write(f'Tabla Correlación para {vari}')
          st.write(a)

          # Variable Independiente
          a1 = a.index.tolist()  # Asegúrate de que a1 sea una lista
          st.write('Variables dependientes:', a1)

          # Declaramos las variables dependientes e independientes para la regresión lineal
          Vars_Indep = data_evaluar2[a1]  # Usa a1 directamente como lista
          Var_Dep = data_evaluar2[vari]

          #Redefinimos las variables
          X = Vars_Indep
          y = Var_Dep

          #Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
          X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= None)

          #Se escalan todos los datos
          escalar = StandardScaler()

          #Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, utilizaremos fit_transform
          X_train = escalar.fit_transform(X_train)
          X_test = escalar.transform(X_test)

          algoritmo = LogisticRegression()

          #Realizamos una prediccion
          y_pred = algoritmo.predict(X_test)
          st.write('Predicción:')
          st.write(y_pred)

          #Verificamos la matriz de confusión
          matriz = confusion_matrix(y_test, y_pred)
          st.write("Matriz de Confusión: ")
          st.write(matriz)

          # Calculo la precision del modelo

          precision = precision_score(y_test, y_pred, average = 'binary', pos_label= 'no')
          st.write('Precisión del modelo: ')
          st.write(precision)

          # Calculo la exactitud del modelo
          exactitud = accuracy_score(y_test,y_pred)
          st.write('Exactitud del modelo: ')
          st.write(exactitud)


          # Calculo la sensibilidad del modelo
          sensibilidad = recall_score(y_test, y_pred, average= 'binary', pos_label = 'yes')
          st.write('Sensibilidad del modelo:')
          st.write(sensibilidad)
      except Exception as e:
        st.write(f"An unexpected error occurred: {e}")

    ###########################################################################################
# Customización del panel general
# Customización del panel general
st.markdown(
   """
    <style>
    .stApp {
        background-image: url('https://images.pexels.com/photos/161212/rio-de-janeiro-olympics-2016-niteroi-brazil-161212.jpeg?cs=srgb&dl=pexels-pixabay-161212.jpg&fm=jpg'); /* Cambia esto por la URL de tu imagen */
        background-size: cover; /* Ajusta la imagen para cubrir todo el fondo */
        background-position: center; /* Centra la imagen */
        background-repeat: no-repeat; /* Evita que la imagen se repita */
    }
    h1 {
        font-size: 40px;
        color: black; /* Cambia el color aquí */
        font-weight: bold;
        text-align: left; /* Alinear a la izquierda el texto */
    }
    h2 {
        font-size: 24px;
        color: black; /* Cambia el color aquí */
        font-weight: bold;
        text-align: left; /* Alinear a la izquierda el texto */
    }
    h3 {
        font-size: 12px;
        color: black; /* Cambia el color aquí */
        font-weight: bold;
        text-align: left; /* Alinear a la izquierda el texto */
    }
    /* Estilo para todos los textos en el body */
    p {
        color: black; /* Cambia el color del texto en párrafos */
    }
    </style>
    """,
    unsafe_allow_html=True
)

```
