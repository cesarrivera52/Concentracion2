```python
import matplotlib.pyplot as plt
```

```python
#Realizamos diagrama de caja o bigote de cada columna del DF
fig= plt.figure(figsize=(15 ,8))
data.plot(kind='box', vert=False)
plt.title('Valores Atípicos de datos SF')
plt.show() #Dibujamos el diagrama
```

```python
y=data
Limite_Superior= y.mean()+ 3*y.std()
Limite_Inferior= y.mean()- 3*y.std()
print('-Límite superior permitido\n', Limite_Superior,'\n')
print('-Límite inferior permitido\n', Limite_Inferior)
```

```python
data2= data[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
valores_nulos= data2.isnull().sum()
valores_nulos
```

```python
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos= data_clean.isnull().sum()
valores_nulos
```

```python
y=data

percentiles25=y.quantile(0.25) #Q1
percentiles75=y.quantile(0.75) #Q3
iqr= percentiles75-percentiles25

Limite_Superior_iqr=percentiles75 + 1.5*iqr
Limite_Inferior_iqr=percentiles25- 1.5*iqr
print("Límite superior permitido: \n",Limite_Superior_iqr)
print("\nLímite inferior permitido: \n",Limite_Inferior_iqr)
```

```python
data2_iqr=data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2_iqr
```

```python
valores_nulos=data2_iqr.isnull().sum()
valores_nulos
```

```python
data_3_iqr= data2_iqr.copy()
data_3_iqr=data_3_iqr.fillna(round(data2_iqr.mean(),1))
```

```python
fig= plt.figure(figsize=(15 ,8))
data_clean.plot(kind='box', vert=False)
plt.title('Valores Atípicos SF Procedimiento "Desviación Estandar" para eliminar outliers en el DF')
plt.show() #Dibujamos el diagrama
```

```python
fig = plt.figure(figsize =(15, 8))
data_3_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos SF del Dataframe")
```
